```python
del cdf['Sno']# or use cdf.drop(['Time','Sno'])
```

```python
del cdf['Time']
```

```python
 cdf=cdf.drop(['ConfirmedIndianNational','ConfirmedForeignNational'],axis=1)
```

```python
cdf
```

```python
cdf['Date']=pd.to_datetime(cdf['Date'],format='%Y-%m-%d') # chanding date column to machine intrepredted date time 
```

```python
cdf
```

```python
cdf['activecu']=cdf['Confirmed'].cumsum()-((cdf['Cured'].cumsum())+(cdf['Deaths'].cumsum()))
cdf['active']=cdf['Confirmed']-(cdf['Cured']+cdf['Deaths'])
```

```python
# cumsum() is used for cumelative sum
cdf
```

```python
statewise=pd.pivot_table(cdf,values=['Confirmed','Deaths','Cured'],index='State/UnionTerritory',aggfunc=max)
```

```python
statewise
```

```python
# recovery rate
statewise['Recovery Rate']=(statewise['Cured']*100)/statewise['Confirmed']
```

```python
statewise
```

```python
# motality rate
statewise['Motality']=(statewise['Deaths']*100)/statewise['Confirmed']
```

```python
statewise
```

```python
t=statewise['Motality'].max()
```

```python
(statewise['Motality']==t).any()
```

```python
statewise=statewise.sort_values(by='Confirmed',ascending=False)
```

```python
statewise_=statewise.style.background_gradient(cmap='cubehelix')
```

```python
statewise_
```

```python
# for worst effected state
result1=cdf.groupby(by='State/UnionTerritory').max()[['active','Date']].sort_values(by=['active'],ascending=False).reset_index()
```

```python
result1
```

```python
fig=mpl.figure(figsize=(18,10))
mpl.title("Top 10 states in india having most active case",size=23)
```

```python
barp=sb.barplot(data=result1.iloc[:10], y='active',x= 'State/UnionTerritory' ,linewidth=2,edgecolor='red')
```

```python
result1=cdf.groupby(by='State/UnionTerritory').max()[['active','Date']].sort_values(by=['active'],ascending=False).reset_index()
fig=mpl.figure(figsize=(18,10))
mpl.title("Top 10 states in india having most active case",size=23)

barp=sb.barplot(data=result1.iloc[:10], y='active',x= 'State/UnionTerritory' ,linewidth=2,edgecolor='red')
mpl.xlabel("State/UT")
mpl.ylabel("Total active cases ")
mpl.show()
```

```python
result2=cdf.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig2=mpl.figure(figsize=(18,10))
mpl.title("Top 10 states in india having most active case",size=23)

barp2=sb.barplot(data=result2.iloc[:10], y='Deaths',x= 'State/UnionTerritory' ,linewidth=2,edgecolor='red')
mpl.xlabel("State/UT")
mpl.ylabel("Total Deaths ")
mpl.show()
```
