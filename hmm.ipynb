```python
pp=p.drop("Cabin",axis=1)
pp.head()
```

```python
e=pd.get_dummies(pp['Pclass'],drop_first=True)
e.head()

```

```python
em=pd.get_dummies(pp['Embarked'],drop_first=True)
em.head()
```

```python
sex=pd.get_dummies(pp['Sex'],drop_first=True)
sex.head()
```

```python
p=pd.concat((pp,em,e,sex),axis=1)
```

```python
p.head()
```

```python
data=p.drop(['Age','Embarked','Name','PassengerId','Ticket','Fare','Sex'],axis=1)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
y= data['Survived']
X=data.drop(['Survived'], axis=1)
X
```

```python
X
```

```python
# Please Do scaling on numerical data to get best result.
from sklearn.preprocessing import StandardScaler
scal=StandardScaler();
X=scal.fit_transform(X)
X
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, 
                                                    random_state=1)
y_train
```

```python
#logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
reg=lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
ypred=lr.predict(X_test)
r1=accuracy_score(y_test,ypred)
print (r1)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression

# Creating an object of Linear Regression
lm = LinearRegression()

# Fit the model using .fit() method
lm.fit(X_train, y_train)
```

```python
# Intercept value
print("Intercept :",lm.intercept_)

# Slope value
print('Slope :',lm.coef_)
```

```python
from sklearn.metrics import r2_score
# Making Predictions of y_value
y_train_pred = lm.predict(X_train)
y_test_pred = lm.predict(X_test)

# Comparing the r2 value of both train and test data
print(r2_score(y_train,y_train_pred))
print(r2_score(y_test,y_test_pred))
```
